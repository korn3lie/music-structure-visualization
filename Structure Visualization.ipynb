```python
import IPython.display as ipd
import lib.util as u

audio_path = 'files/FMP/FMP_C4_Audio_Brahms_HungarianDances-05_Ormandy.mp3'
segment_partitions, SSM_len, audio_duration = u.partition_audio_into_similarity_segments(audio_path, min_segment_len_seconds = 4, penalty = -2, thresh_chroma=0.3, thresh_ssm=0.7)
pixelArt = u.visualize_segment_partitions(segment_partitions, SSM_len, show = True)
```

```python
fps_video = pixelArt.shape[1] / audio_duration

# output dimensions of the video:
width = 1920 
height = 200
output_path = 'result2.mp4'

frames = u.create_frames(pixelArt, audio_duration, fps_video, width, height)
u.create_video_with_audio(frames, audio_path, output_path, fps_video)
```
